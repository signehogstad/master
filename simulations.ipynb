```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def signal_intensity(theta, PD, TR,  TE, T1, T2):
    theta = theta * np.pi / 180
    E1 = np.exp(-TR / T1)
    E2 = np.exp(-TE / T2)
    return PD * np.sin(theta) * ((1 - E1) / (1 - (np.cos(theta) * E1))) * E2
```

```python
TR = 29
TE1 = 7.5
TE2 = 15
TE3 = 22.5
PD_GM = 0.84
PD_WM = 0.68
T1_GM = 1600
T1_WM = 900
T2_GM = 66
T2_WM = 50
```

```python
thetas = np.linspace(1, 90, 100)
```

```python
S_WM_TE1 = [signal_intensity(theta, PD_WM, TR, TE1, T1_WM, T2_WM) for theta in thetas]
S_WM_TE2 = [signal_intensity(theta, PD_WM, TR, TE2, T1_WM, T2_WM) for theta in thetas]
S_WM_TE3 = [signal_intensity(theta, PD_WM, TR, TE3, T1_WM, T2_WM) for theta in thetas]
S_GM_TE1 = [signal_intensity(theta, PD_GM, TR, TE1, T1_GM, T2_GM) for theta in thetas]
S_GM_TE2 = [signal_intensity(theta, PD_GM, TR, TE2, T1_GM, T2_GM) for theta in thetas]
S_GM_TE3 = [signal_intensity(theta, PD_GM, TR, TE3, T1_GM, T2_GM) for theta in thetas]
```

```python
print(signal_intensity(20, PD_WM, TR, TE1, T1_WM, T2_WM))
```

```python
plt.plot(thetas, S_WM_TE1, "--", c="k", label="$S_{WM}$")
#plt.plot(thetas, S_WM_TE2, "--", c="tab:blue")
#plt.plot(thetas, S_WM_TE3, "--", c="tab:orange")
plt.plot(thetas, S_GM_TE1, c="k", label="$S_{GM}$")
#plt.plot(thetas, S_GM_TE2, c="tab:blue")
#plt.plot(thetas, S_GM_TE3, c="tab:orange")
# plt.axvline(x = 6, color = 'k', label = 'axvline - full height')
# plt.axvline(x = 27, color = 'k', label = 'axvline - full height')
plt.xlabel(r"FA ($\theta$)")
plt.ylabel("Signal intensity")
plt.legend()
```
