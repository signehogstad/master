```python
import sys
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
import os
midbrain_folder = "../Figures/midbrain_00001573/"
brain_folder = "../Figures/brain_nomask_00001573/"
sys.path.append(midbrain_folder)
```

```python
fig, axes = plt.subplots(8, 5, figsize=(15,25), sharey=True)
for path, directories, files in os.walk(sys.path[-1]):
    #files.remove(".DS_Store")
    for file in files:
        img = mpimg.imread(midbrain_folder + f"/{file}")
        index = files.index(file)
        j = index//5
        i = index%5
        axes[j,i].imshow(img)
        axes[j,i].axis('off')
        axes[j,i].set_title(file[:-22])
        
axes[7,3].axis('off')
axes[7,4].axis('off')
```

```python
fig, axes = plt.subplots(8, 5, figsize=(12,25), sharey=True)
index = 0
for path, directories, files in os.walk(sys.path[-1]):
    #files.remove(".DS_Store")
    for file in files:
        img = mpimg.imread(brain_folder + f"/{file}")
        j = index//5
        i = index%5
        axes[j,i].imshow(img[:590, 140:635, :])
        axes[j,i].axis('off')
        name = file[:-26]

        if "_e3" in name:
            name = name[:-3]
        if "_A2" in name:
            name = name[:-3]
        if "_ECHO-3" in name:
            name = name[:-7]
        if name[-1] == "_":
            name = name[:-1]
        if name[-1] == "a":
            name = name[:-1]

        axes[j,i].set_title(name)
        index += 1

        
axes[7,3].axis('off')
axes[7,4].axis('off')
plt.subplots_adjust(hspace=-0.3)
```
